**Now, let's read in the sessions dataset!**

```r
url_sessions <- "https://raw.githubusercontent.com/tchan0717/dsci-100-2025w1-group-36/refs/heads/main/data/sessions.csv"
sessions_data <- read_csv(url_sessions)
sessions_data
```

#### Description of the set:
Indicated at the top of the dataframe, "sessions.csv" has 5 variables and 1535 observations (1535x5). The variables, their data type, and descriptions are listed below: 
1. `hashedEmail`: *chr (character)* - player's email in hashed form
2. `start_time`: *chr (character)* - player's game session start time in "dd/mm/yyyy" and "time (in 24 hour interval)"
4. `end_time`: *chr (character)* - player's game session end time in "dd/mm/yyyy" and "time (in 24 hour interval)"
5. `original_start_time`: *dbl (double (decimal))* - players' start time in UNIX format (milliseconds)
6. `original_end_time`: *dbl (double (decimal))* - players' end time in UNIX format (milliseconds)


The only issue is the start_time and end_time columns. They are in a character data type, which is going to make it difficult to perform functions on if we wanted to use the times and treat them as numerical forms. There are the original_start_time and original_end_time variables which represent the start_time and end_time variables in a different format that is numeric. But, they are difficult to interpret and mutate.

The data is already tidy, but before we summarize, I will wrangle the data by separating the start_time and end_time column values so that I can change the time into a decimal data type in hours, and then make a new column for start time in hours as an integer. This will make it easier to use for later.

```r
sessions_data_tidy <- sessions_data |>
                        separate(end_time, 
                                into = c("date_end", "end_time"),
                                sep = " ") |>
                        separate(start_time,
                                into = c("date_start", "start_time"),
                                sep = " ") |>
                        mutate(start_time = as.numeric(hm(start_time))/3600) |> #lubridate's hm function extracts the time in seconds
                        mutate(end_time = as.numeric(hm(end_time))/3600) |> #devide it by 3600 to get time in hours
                        mutate(start_time_hr = as.integer(start_time)) |>
                        mutate(start_time_hr = start_time_hr %% 24) #%% will divide each value by 24 and output the remainder; we want this so that any 24 hour values = 0 on the 24 hour interval
sessions_data_tidy
```

### 4. Methods and Plans

Choosing an appropriate prediction model is very important to ensure an accurate and strong performance. I would approach my question using non-linear KNN regression. Briefly, I would work with the players.csv dataset as it contains relevant variables I need; played_hours, Age, and I can mutate the experience column to make each category assigned to a numerical value from levels of 1 - 5. Then, I will do necessary training and validation of the data to create an optimal KNN non-linear regression model. 

KNN regression is appropriate because I want to predict a numerical value, played_hours. To compare, KNN classification is for categorical predictions, not numerical. Judging previous visualizations, it doesn't seem like the variables have a linear relationship with each other, so linear regression is not preferred. Using the explanatory and response variable chosen, the goal is to efficiently input any age and experience and the classifier will output a prediction of how many total hours they may contribute. Then, it can be decided to target those players or not for future session data collections.

I chose these explanatory variables because they are descriptions of "kinds" of players and I thought they would provide the most meaningful and varying data to explore and predict from. Typically, they contribute a lot to determining how long one plays a game. Gender was considered too. I felt like gender was unreliable to predict from because there are some "prefer not to say" values which will not be helpful in narrowing down which gender to possibly "target". I didn't choose any of the time variables because I felt like it doesn't correlate well to "kinds" of players, and each players' session times start and end randomly, it is not a unique behaviour to each player like age and experience are.

##### *To process the data, I will follow the general non-linear knn regression model creation steps and make notes of specific changes where necessary:*
1. Make new column for experience in a numerical level format (ex. 1 = beginner, 5 = pro)
   - *ex. mutate(experience_levels = as.numeric(factor(experience, levels = c("Beginner", "Amateur", "Regular", "Veteran", "Pro"))))*
2. Inspect data and handle missing values and problems appropriately
3. Split dataset
    - 75% training and 25% testing.
4. Tune the training set
    - In knn_spec: weight_func = "rectangular", set mode and engines to "regression" and "kknn"
    - In recipe: response variable = total_hours, explanatory variable = Age + experience_levels
        - Scale with *step_scale* and *step_center*
5. Cross validation on training set
        - 5 folds
    - Make a tibble with range of neighbors from 1 to 196
        - Neighbors differ by 4; ex. neighbors = 1, 5,... 196
    - Create workflow and use *tune_grid* and *collect_metrics*
6. Filter and slice for optimal *k*
7. Assess on testing data by making new knn model and workflow, then using *predict* and *bind_cols*, and collect metrics
8. Compare RMSPE to cross-validation RMSE
9. Use optimal *k* and fit into original dataset to create final model

To conclude, fortunately, KNN regression doesn't require much assumptions compared to linear regression assuming linearity. Though, it does assume that new observations are similar to data it was trained with. Thus, one setback is that its predictions can be inaccurate when predicting observations outside the range of its training data. Moreover, the larger the dataset, the longer it takes to compute the output. Luckily, this dataset is not too large, so computing should not be an issue. However, it can become a problem for when more data is added. It's also sensitive to noisy data and because it predicts by distance, scaling is needed to ensure one variable doesn't overtake the other.
